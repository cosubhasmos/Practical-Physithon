STATEMENT : TO FIND THE THERMAL CONDUCTIVITY OF A BAD CONDUCTOR BY LEE'S METHOD.

The thermal conductivity of bad conductor $(k)$ is calculated by following formulae,

$$k = \dfrac{msd}{A(Q_1 - Q_2)}*\dfrac{d\theta}{dt}$$

In which,
$\dfrac{d\theta}{dt}$ is to be calculated from the graph. Hence, here we are going to fit those observational data into plot to get slope, which will help us to find thermal conductivity of bad conductor. 


Let's import some necessary libraries first.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.interpolate import interp1d
```

```python
fig, ax = plt.subplots(figsize=(12,8))
```

Now its time to fit the observed data from experiment. You may have measured the tempreture in several time interval and noted out the tempr in that minute. In example below, it is clear that tempreture is observed in 0, 5, 10, 14, 18, and 20th minute during cooling and respective observations of tempreture at that minute is fitted in tempr array respectively.

```python
time = [0,5,10,14,18,20]
tempr = [85.4,74.5,67.3,63,58.7,56.7] 
```

To acquire beautiful curve, we must ignore pointy curve, which can be done by interpolating method.

```python
f = interp1d(time,tempr,kind="cubic")
x_new = np.linspace(0, 20, num=50, endpoint=True)
y_new = f(x_new)
```

It is necessary to show intersecting or noted values in graph to make more sense about the curve, which can easily done by annotating them.

```python
for i, (x, y) in enumerate(zip(time, tempr)):
    ax.annotate(f"({x}, {y})", (x, y), textcoords="offset points", xytext=(0, 10), ha='center')
```

Since, everything is set, its time to plot the curve.

```python
ax.plot(x_new,y_new)
ax.annotate(f"(5,67.3)", (5, 67.3), textcoords="offset points", xytext=(0, 10), ha='center')
ax.grid("science")
ax.set_title("Linear Graph between Time and Temperature during Cooling")
ax.set_xlabel("Time $(t)s$",fontsize=15)
ax.set_ylabel("Temperature $(T)^{o C}$",fontsize=15)
ax.plot([5,5],[74.5,67.3])
ax.plot([5,10],[67.3,67.3])
```

It is essential to determine the slope to calculate the efficiency of the conductor, which is done by following simple process.

```python
x_point = 5
h = 0.001
slope = (74.5-67.3) / (5)
print("Slope:%.2f"%slope)
```
